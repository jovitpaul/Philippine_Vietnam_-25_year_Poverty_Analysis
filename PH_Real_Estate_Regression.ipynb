```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset
df = pd.DataFrame()

try:
    # TRY TO LOAD YOUR KAGGLE DATA HERE:
    df = pd.read_csv('Metro_Manila_Properties.csv')
    print("Actual Kaggle Data Loaded Successfully!")
except FileNotFoundError:
    print("CSV not found. Please check the spelling of your filename in the left folder menu.")

# Let's look at the first 5 properties
display(df.head())
```

```python
# --- IMPORTANT: Change these to match the exact column names in your Kaggle data! ---
# For example, they might be 'Area_sqm' and 'Price_PHP'
X_column = 'Floor_Area '
Y_column = 'Price'

# Prepare the data
# We drop missing values to ensure the math works perfectly
clean_df = df.dropna(subset=[X_column, Y_column])

X = clean_df[[X_column]] # The features (Predictor)
Y = clean_df[Y_column]   # The target (What we want to predict)

# Initialize the Linear Regression Model
model = LinearRegression()

# Train (Fit) the model using the actual data
model.fit(X, Y)

# Get the mathematical formula (y = mx + b)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Regression Formula: Price = ({slope:,.2f} * Floor Area) + {intercept:,.2f}")
print(f"Insight: For every 1 extra square meter, the price increases by PHP {slope:,.2f}")
```

```python
plt.figure(figsize=(10, 6))

# Plot the actual properties as scatter dots
plt.scatter(X, Y, color='blue', alpha=0.5, label='Actual Properties')

# Plot the model's Regression Line
plt.plot(X, model.predict(X), color='red', linewidth=3, label='Regression Line')

# Format the graph
plt.title('Metro Manila Property Prices vs. Floor Area', fontsize=16)
plt.xlabel('Floor Area (Square Meters)', fontsize=12)
plt.ylabel('Price (PHP)', fontsize=12)

# This formats the Y-axis to show regular numbers instead of scientific notation (e.g. 1e7)
plt.ticklabel_format(style='plain', axis='y')

plt.legend()
plt.grid(True)
plt.show()
```
